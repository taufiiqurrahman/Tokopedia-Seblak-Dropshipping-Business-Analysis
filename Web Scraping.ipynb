# A. Web Scraping

Pada bagian A dilakukan pengambilan data dari halaman pencarian kata kunci produk "seblak" di Tokopedia.

```python
#import module
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
```

```python
#Assign driver
driver = webdriver.Chrome ()
```

```python
#List Kosong buat Value
Nama = []
Harga = []
Penjual = []
Kota = []
Terjual = []
Rate = []

#loop dari hal 1 sampai 10
for i in range (1,11):
    #Akses ke halaman situs
    site = f'https://www.tokopedia.com/search?navsource=&page={i}&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st='
    driver.get(site)

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

#ngambil kotak kotak_kecil
    kotak_kecil = soup.find_all('div',{"class":"css-llwpbs"})

    for satu_kotak in kotak_kecil[:10]:

        #Mengambil Nama
        try:
            nama_produk = satu_kotak.find('div',{"class":"prd_link-product-name css-3um8ox"}).get_text()
            Nama.append(nama_produk)

        except:
            nama_produk = "kosong"
            # print('nama_produk', nama_produk)
            Nama.append(nama_produk)

        #Ekstrak Penjual
        try:
            nama_toko = satu_kotak.find('span',{"class":"prd_link-shop-name css-1kdc32b flip"}).get_text()
            Penjual.append(nama_toko)

        except:
            nama_toko = 'kosong'
            Penjual.append(nama_toko)

        #Kota
        try:
            kota_toko = satu_kotak.find('span',{"class":"prd_link-shop-loc css-1kdc32b flip"}).get_text()
            Kota.append(kota_toko)

        except:
            kota_toko = 'kosong'
            Kota.append(kota_toko)

        #Harga
        try:
            harga_produk = satu_kotak.find('div',{"class":"prd_link-product-price css-h66vau"}).get_text()
            Harga.append(harga_produk)

        except:
            harga_produk = "kosong"
            # print('nama_produk', nama_produk)
            Harga.append(harga_produk)

        #Terjual
        try:
            produk_terjual = satu_kotak.find('span',{"class":"prd_label-integrity css-1sgek4h"}).get_text()
            Terjual.append(produk_terjual)

        except:
            produk_terjual = None
            Terjual.append(produk_terjual)

        #Rate
        try:
            rating_terjual = satu_kotak.find('span',{"class":"prd_rating-average-text css-t70v7i"}).get_text()
            Rate.append(rating_terjual)

        except:
            rating_terjual = None
            Rate.append(rating_terjual)

#untuk menampilkan tabel beserta kolomnya
df = pd.DataFrame({"nama_produk" : Nama,
                   "nama toko" : Penjual,
                   "kota" : Kota,
                   "harga" : Harga,
                   "terjual" : Terjual,
                   "rating" : Rate})

df
        

```

```python
#untuk elihat value dan type data setiap kolom
df.info()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Menghitung korelasi antara harga dan rating
corr_r, pval_p = stats.pearsonr(df['harga'], df['rating'])
corr_rho, pval_s = stats.spearmanr(df['harga'], df['rating'])
corr_tau, pval_k = stats.kendalltau(df['harga'], df['rating'])

# Menampilkan hasil korelasi
print("\nKorelasi antara Harga dan Rating:")
print(f"r-Korelasi: {corr_r:.2f}, p-value: {pval_p}")
print(f"rho-Korelasi: {corr_rho:.2f}, p-value: {pval_s}")
print(f"tau-Korelasi: {corr_tau:.2f}, p-value: {pval_k}")
```
